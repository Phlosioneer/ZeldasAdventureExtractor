```python
class Actor:
    def __init__(self, stream: StructStream):
        assert len(stream) == 54, len(stream)
        
        pointers = stream.takeRaw(4 * 6)
        assert pointers == b'\0' * (4 * 6), pointers

        self.direction, self.type, self.unk_0x1c, self.descIndex = stream.take("HHbB")
        y, x, self.health, self.unk_0x24, self.isBeingUsed_maybe = stream.take("hhHHH")
        self.unk_0x1e = {"x": x, "y": y}

        self.unk_0x28, self.unk_0x2a, y, x, self.unk_0x30 = stream.take("HHHHB")
        self.unk_0x31, self.unk_0x32, self.unk_0x34, self.unk_0x35 = stream.take("BHBB")
        assert len(stream) == 0, stream

        self.description: Optional[ActorDescription] = None

    def toPlainDict(self) -> dict:
        ret = copy.copy(self.__dict__)
        del ret["description"]
        return ret

class ActorDescription:
    def __init__(self, stream: StructStream):
        assert len(stream) == 46

        y, x, self.groupCount, pointer = stream.take("hhHI")
        self.size = {"x": x, "y": y}
        assert pointer == 0, pointer

        self.unk_0xa, self.unk_0xe, self.unk_0xf, self.useCost = stream.take("IBBH")
        self.unk_0x12, self.unk_0x14, self.unk_0x15 = stream.take("HBB")
        self.unk_0x16 = []
        for _ in range(3):
            y, x = stream.take("hh")
            self.unk_0x16.append({"x": x, "y": y})
        self.unk_0x22, self.unk_0x24, self.unk_0x26 = stream.take("HHH")
        self.interactsWithItem = stream.take("H")
        self.unk_0x2a, self.unk_0x2b, self.unk_0x2c = stream.take("BBH")
        assert len(stream) == 0, stream

        self.groups: Optional[List[SpriteGroup]] = None
        self.scripts: Optional[dict]  = None
        self.unknownFieldFromSpriteTree: Optional[int] = None
        self.extraUnusedGroups: Optional[List[SpriteGroup]] = None
        self._cachedHashOfGroups: Optional[int] = None

    def toPlainDict(self) -> dict:
        if len(self.groups) > 0:
            assert isinstance(self.groups[0], SpriteGroup)
        
        ret = copy.copy(self.__dict__)
        ret["groups"] = [g.toPlainDict() for g in self.groups]
        ret["unkMulti_0x1a"] = list(self.unkMulti_0x1a)
        return ret
    
    def hashOfSpriteGroups(self) -> int:
        if self._cachedHashOfGroups == None:
            self._cachedHashOfGroups = hash(tuple([g.hashOfSprites() for g in self.groups]))
        return self._cachedHashOfGroups

class SpriteGroup:
    def __init__(self, stream: StructStream):
        assert len(stream) >= 72

        self.animationFrameOrder = list(stream.take("16b"))
        frameCount, self.unk_0x12, self.unk_0x14 = stream.take("HHH")
        polyVertexCount = stream.take("H")
        self.unk_0x18 = []
        for _ in range(3):
            y, x = stream.take("hh")
            self.unk_0x18.append({"x": x, "y": y})
        self.boundingPolygon = []
        for _ in range(8):
            y, x = stream.take("hh")
            self.boundingPolygon.append({"x": x, "y": y})
        pointer = stream.take("I")
        assert pointer == 0, pointer

        self.animationFrameOrder = self.animationFrameOrder[:frameCount]
        self.boundingPolygon = self.boundingPolygon[:polyVertexCount]
        assert len(stream) == 0, stream

        self.sprites: Optional[List[PIL.Image.Image]] = None
        self.unknownFieldFromSpriteTree: Optional[int] = None
        self._cachedHash: Optional[int] = None
    
    def toPlainDict(self) -> dict:
        ret = copy.copy(self.__dict__)
        del ret["sprites"]
        return ret
    
    def hashOfSprites(self) -> int:
        if self._cachedHash == None:
            self._cachedHash = hash(tuple([hash(s.tobytes()) for s in self.sprites]))
        return self._cachedHash

def parseActors(data, prefix = "") -> Tuple[List[Actor], List[ActorDescription]]:
    tree = ResourceTree.parseFromStream(StructStream(data, endianPrefix=">"))
    castLabel = prefix + "sp_cast"
    descLabel = prefix + "sp_desc"
    groupLabel = prefix + "sp_groups"
    assert castLabel in tree.children, tree.children
    actors = [Actor(s) for s in tree.children[castLabel].elements]

    if len(actors) != 0:
        assert descLabel in tree.children, tree.children
        assert groupLabel in tree.children, tree.children
    
    descriptions = []
    if descLabel in tree.children:
        descriptions = [ActorDescription(s) for s in tree.children[descLabel].elements]
        groups = [SpriteGroup(s) for s in tree.children[groupLabel].elements]

        groupIndex = 0
        for actor in actors:
            actor.description = descriptions[actor.descIndex]
        for desc in descriptions:
            desc.groups = groups[groupIndex:groupIndex + desc.groupCount]
            groupIndex += desc.groupCount
    
    return actors, descriptions
"""
display(Actor(commonResources.children["zsp_cast"].elements[0].copy()).__dict__)
display(ActorDescription(commonResources.children["zsp_desc"].elements[0].copy()).__dict__)
for s in commonResources.children["zsp_groups"].elements:
    display(SpriteGroup(s.copy()).__dict__)
"""

zeldaActor, _ = parseActors(mainFile.subFiles["zinit"].getRecord(0, kind="data"), prefix="z")
zeldaActor[0].description.__dict__
```

```python
def decompressSprite(stream: StructStream, palette):
    lineCount = stream.take("I")
    stream = stream.takeFork(lineCount)

    #print(stream)

    pixels = b''
    while stream.peekRaw(4) != b'\0\0\0\0':
        skip, size = stream.take("HH")
        pixels += b'\0' * skip
        pixels += stream.takeRaw(size * 4)

    if len(pixels) == 0:
        # Blank image.
        return PIL.Image.frombytes("RGBA", (1, 1), b'\0\0\0\0')

    partialRow = len(pixels) % 384
    pixels += b'\0' * (384 - partialRow)


    rows = [pixels[i * 384:(i + 1) * 384] for i in range(len(pixels)//384)]
    maxWidth = 0
    for row in rows:
        for i, b in enumerate(row):
            if b != 0:
                maxWidth = max(maxWidth, i)

    trimmedRows = [row[:maxWidth] for row in rows]
    pixels = b''.join(trimmedRows)

    #print(maxWidth, len(rows), len(pixels))

    img = PIL.Image.frombytes("P", (maxWidth, len(rows)), pixels)
    img.putpalette(palette)
    return img

def makeRandomClut() -> bytearray:
    import random
    return bytearray([random.randint(0, 255) for _ in range(256 * 3)])
```

```python

def unpackPointerArray(stream: StructStream) -> Tuple[List[StructStream], int]:
    length, unknownField = stream.take("II")
    #assert reserved == 0, reserved
    #if reserved != 0:
    #    print("Warning: reserved = {:08X}".format(reserved))
    
    offsets = [stream.take("I") for _ in range(length)]
    elements = [stream.takeFork(offsets[i + 1] - offsets[i]) for i in range(len(offsets) - 1)]
    elements.append(stream.fork())
    return elements, unknownField

def unpackSpriteTree(actorDescriptions: List[ActorDescription], data: bytes, palette):
    topStream = StructStream(data, endianPrefix=">")
    actorTrees, unknownField = unpackPointerArray(topStream)
    assert len(actorDescriptions) <= len(actorTrees), "{} <= {}".format(len(actorDescriptions), len(actorTrees))

    

    for desc, actorTree in zip(actorDescriptions, actorTrees):
        groupTrees, unknownField = unpackPointerArray(actorTree)
        desc.unknownFieldFromSpriteTree = unknownField

        groups = desc.groups
        assert len(groups) <= len(groupTrees), "{} <= {}".format(len(groups), len(groupTrees))
        
        for group, groupTree in zip(groups, groupTrees):
            sprites, unknownField = unpackPointerArray(groupTree)
            group.unknownFieldFromSpriteTree = unknownField
            group.sprites = [decompressSprite(s, palette) for s in sprites]
        
        extraGroups = []
        for groupTree in groupTrees[len(groups):]:
            sprites, _ = unpackPointerArray(groupTree)
            extraGroups.append([decompressSprite(s, palette) for s in sprites])
        if len(extraGroups) > 0:
            desc.extraUnusedGroups = extraGroups
    
    unusedGroups = []
    for actorTree in actorTrees[len(actorDescriptions):]:
        groupTrees, _ = unpackPointerArray(actorTree)
        for groupTree in groupTrees:
            sprites, _ = unpackPointerArray(groupTree)
            unusedGroups.append([decompressSprite(s, palette) for s in sprites])
        
    return unusedGroups

cellName = "j24"
display(dyuvToRGBBackground(overFiles.subFiles[cellName].getRecord(0, kind="video")))
cellActors, cellDescriptions = parseActors(overFiles.subFiles[cellName].getRecord(4, kind="data"))
cellPalette = getClut(overFiles.subFiles[cellName].getRecord(7, kind="data"))
cellSprites = overFiles.subFiles[cellName].getRecord(5, kind="data")
unpackSpriteTree(cellDescriptions, cellSprites, cellPalette)
for i, actor in enumerate(cellActors):
    for j, group in enumerate(actor.description.groups):
        for k, sprite in enumerate(group.sprites):
            print("Actor", i, "Group", j, "Sprite", k)
            display(sprite)
```

```python
def rl7ToRGB(data, palette:bytes, transparentColor = None, emptySpaceColorIndex = 0, forceWidth = None):
    assert isinstance(palette, bytes)
    lastNonzeroIndex = 0
    for i, b in enumerate(data):
        if b != 0:
            lastNonzeroIndex = i
    #print(lastNonzeroIndex, len(data))
    truncImageData = data[:lastNonzeroIndex + 2]

    i = 0
    pixelRows = []
    currentRow = b''
    while i < len(truncImageData):
        if truncImageData[i] & 0x80 == 0:
            #print("single")
            currentRow += truncImageData[i].to_bytes(1)
            i += 1
        elif truncImageData[i + 1] == 0:
            pixelRows.append(currentRow)
            currentRow = b''
            i += 2
        else:
            c = (truncImageData[i] & 0x7F).to_bytes(1)
            for _ in range(truncImageData[i + 1]):
                currentRow += c
            i += 2
    if len(currentRow) != 0:
        print("Warning: last row had data")
        pixelRows.append(currentRow)
    del currentRow
    del i

    if forceWidth == None:
        width = max([len(r) for r in pixelRows])
    else:
        width = forceWidth
    height = len(pixelRows)
    pixels = b''
    for row in pixelRows:
        pixels += row 
        for _ in range(width - len(row)):
            pixels += emptySpaceColorIndex.to_bytes(1)
    assert len(pixels) == width * height

    if len(pixels) == 0:
        print("Empty frame")
        ret = PIL.Image.frombytes('P', (1, 1), b'\0')
        ret.putpalette(b'\0\0\0\0', rawmode = "RGBA")
        return ret

    img = PIL.Image.frombytes('P', (width, height), pixels)

    
    rgbaPalette = b''
    i = 0
    while i < len(palette):
        c = palette[i:i + 3]
        rgbaPalette += c
        if c == transparentColor:
            rgbaPalette += b'\0'
        else:
            rgbaPalette += 0xFF.to_bytes(1)
        i += 3
    img.putpalette(rgbaPalette, rawmode="RGBA")
    return img

cellUnkImage = overFiles.subFiles["g23"].getRecord(1, kind="video")
display(rl7ToRGB(cellUnkImage, bytes(makeRandomClut()), emptySpaceColorIndex=4))
```

```python
class Script:
    def __init__(self, conditionArray, onTrueArray, onFalseArray):
        self.conditions = [ScriptCondition(s.take("I")) for s in conditionArray.elements]
        self.onTrue = [ScriptAction(s.take("I")) for s in onTrueArray.elements]
        self.onFalse = [ScriptAction(s.take("I")) for s in onFalseArray.elements]
    
    def __repr__(self) -> str:
        return "<If {} then {} else {}>".format(self.conditions, self.onTrue, self.onFalse)

# Formatted (name, usesIndex, usesParam)
OPCODE_LOOKUP = {
    0: ("nop", False, False),
    1: ("assign", True, True),
    2: ("add", True, True),
    3: ("subtract", True, True),
    4: ("loseOneHeart", False, False),
    7: ("showSparkles_maybe", False, False),
    9: ("setPlayerX", False, True),
    10: ("setPlayerY", False, True),
    14: ("setSpriteGroup", False, True),
    17: ("registerAsTreasureListener", False, False),
    20: ("darkenRoom_maybe", False, False),
    23: ("setIsVisible", False, True),
    25: ("setCellCoords", True, True),
    26: ("playSound3", False, False),
}
class ScriptAction:
    def __init__(self, code):
        opcode = (code & 0xFC00_0000) >> 26
        self.index = (code & 0x03FF_0000) >> 16
        self.parameter = (code & 0xFFFF)

        self.usesIndex = False
        self.usesParameter = True

        if opcode in OPCODE_LOOKUP:
            self.opcode, self.usesIndex, self.usesParameter = \
                OPCODE_LOOKUP[opcode]
        elif opcode == 5:
            if self.index > 0:
                self.opcode = "addOneMaxHealth"
                self.usesParameter = False
            elif self.parameter == 0:
                self.opcode = "saveMaxHealthToUnkVar"
                self.usesParameter = False
            else:
                self.opcode = "setUnkHealthVar"
        elif opcode == 11:
            if self.index > 0:
                self.opcode = "setOutputVarAndPlaySound2"
            else:
                self.opcode = "setOutputVar"
        elif opcode == 19:
            if self.index > 0:
                self.opcode = "addRupees"
            else:
                self.opcode = "removeRupees"
        else:
            self.opcode = "op{}".format(opcode)
            self.usesIndex = True

    def __repr__(self) -> str:
        if self.usesIndex and self.usesParameter:
            return "{}({}, {})".format(self.opcode, self.index, self.parameter)
        elif self.usesIndex:
            return "{}({})".format(self.opcode, self.index)
        elif self.usesParameter:
            return "{}({})".format(self.opcode, self.parameter)
        return self.opcode

CONDITION_LOOKUP = [
    "always",
    "save[{}] <= {}",
    "save[{}] > {}",
    "save[{}] != {}",
    "save[{}] == {}",
    "always (corrupted)",
    "always (corrupted)"
]
class ScriptCondition:
    def __init__(self, code):
        self.opcode = (code & 0xFC00_0000) >> 26
        self.index = (code & 0x03FF_0000) >> 16
        self.parameter = (code & 0xFFFF)
    def __repr__(self) -> str:
        if self.opcode >= len(CONDITION_LOOKUP):
            return "always ({})".format(self.opcode)
        ret = CONDITION_LOOKUP[self.opcode]
        if "{" in ret:
            return ret.format(self.index, self.parameter)
        else:
            return ret

ACTOR_SCRIPT_TYPE_LOOKUP = {
    "default": "actorType{}",
    1: "onTouch_maybe",
}
CELL_SCRIPT_TYPE_LOOKUP = {
    "default": "cellType{}",
}
def parseScriptTypes(tree, typeNameLookupTable):
    allScripts = {}
    for i, subTree in enumerate(tree.children.values()):
        if i in typeNameLookupTable:
            kind = typeNameLookupTable[i]
        else:
            kind = typeNameLookupTable["default"].format(i)

        allScripts[kind] = parseScriptPseudoArray(subTree)
    return allScripts

def parseScriptPseudoArray(root):
    scripts = []
    assert len(root.children) % 3 == 0
    count = len(root.children) // 3
    for i in range(count):
        scriptTrees = [
            root.children[i * 3], 
            root.children[i * 3 + 1],
            root.children[i * 3 + 2]
        ]
        script = Script(scriptTrees[0], scriptTrees[1], scriptTrees[2])
        scripts.append(script)
    return scripts

class CycleDefinition:
    CYCLE_MODES = {
        1: "IncreaseOnly",
        2: "Oscillate"
    }

    CYCLE_DIRECTIONS = {
        1: "Increasing",
        2: "Decreasing"
    }

    STAGGER_MODES = {
        0: "None",
        3: "RandomAfterLooping",
    }

    def __init__(self, stream):
        assert len(stream) == 20, len(stream)
        self.start, self.length, mode, self.delay = stream.take("HHHH")
        self.currentOffset, cycleDirection, enabled, self.timer = stream.take("HHHh")
        staggerMode, hasLooped = stream.take("HH")
        
        assert len(stream) == 0, stream

        if mode in CycleDefinition.CYCLE_MODES:
            self.mode = CycleDefinition.CYCLE_MODES[mode]
        else:
            print("Unknown cycle mode", mode)
            self.mode = "UnknownMode{}".format(mode)
        
        if cycleDirection in CycleDefinition.CYCLE_DIRECTIONS:
            self.cycleDirection = CycleDefinition.CYCLE_DIRECTIONS[cycleDirection]
        else:
            print("Unknown cycle direction", cycleDirection)
            self.cycleDirection = "UnknownDirection{}".format(cycleDirection)
        
        if staggerMode in CycleDefinition.STAGGER_MODES:
            self.staggerMode = CycleDefinition.STAGGER_MODES[staggerMode]
        else:
            print("Unknown stagger mode", staggerMode)
            self.staggerMode = "UnknownStaggerMode{}".format(staggerMode)
        
        self.enabled = enabled != 0
        self.hasLooped = hasLooped != 0

class CellInfo:
    def __init__(self, data):
        stream = StructStream(data, endianPrefix=">")
        tree = ResourceTree.parseFromStream(stream)
        
        self.infoUnk0, hasSprites, hasPaletteCycling = tree.children["info"].elements[0].peek("HHH")
        self.hasSprites = hasSprites != 0
        self.hasPaletteCycling = hasPaletteCycling != 0

        self.cyclers = [CycleDefinition(s) for s in tree.children["cycle"].elements]

        if len(tree.children["play;"].elements) > 0:
            self.music = tree.children["play;"].elements[0].takeNullTermString().decode('ascii')

        #display(tree.children["voice"].__dict__)

class Cell:
    def __init__(self, subFile: zelda_adventure_filesystem.ResourceFileSystemFolder, name: str):
        self.name = name

        self.info = CellInfo(subFile.getRecord(2, kind="data"))
        self._parseBackground(subFile)
        self._parseActors(subFile)
        self._parseScripts(subFile)
        self._parseCollisionData(subFile)

    def _parseActors(self, subFile: zelda_adventure_filesystem.ResourceFileSystemFolder):
        self.actors, self.descriptions = parseActors(subFile.getRecord(4, kind="data"))
        self.palette = getClut(subFile.getRecord(7, kind="data"))
        
        if len(self.descriptions) > 0:
            assert isinstance(self.descriptions[0], ActorDescription), self.descriptions

        sprites = subFile.getRecord(5, kind="data")
        hasNonzeroByte = False
        for b in sprites:
            if b != 0:
                hasNonzeroByte = True
                break
        
        if hasNonzeroByte and not self.info.hasSprites:
            print("hasSprites false when sprite is present. Cell:", self.name)
        if hasNonzeroByte:
            self.unusedSprites = unpackSpriteTree(self.descriptions, sprites, self.palette)
        else:
            self.unusedSprites = []

    def _parseScripts(self, subFile: zelda_adventure_filesystem.ResourceFileSystemFolder):
        scriptFile = subFile.getRecord(6, kind="data")
        scriptFileTree = ResourceTree.parseFromStream(StructStream(scriptFile, endianPrefix=">"))
        for desc, tree in zip(self.descriptions, scriptFileTree.children.values()):
            desc.scripts = parseScriptTypes(tree, ACTOR_SCRIPT_TYPE_LOOKUP)

        cellScriptTree = scriptFileTree.children[len(self.descriptions)]
        self.scripts = parseScriptTypes(cellScriptTree, CELL_SCRIPT_TYPE_LOOKUP)

        cellVarArray = scriptFileTree.children[len(self.descriptions) + 1]
        if self.name != "gl6":
            self.vars = [s.take("H") for s in cellVarArray.children[0].elements]
            lastUsedIndex = len(self.descriptions) + 1
        else:
            self.vars = None
            lastUsedIndex = len(self.descriptions)
        
        self.extraScriptData = [scriptFileTree.children[i].simplify() for i in \
                range(lastUsedIndex + 1, len(scriptFileTree.children))]

    def _parseBackground(self, subFile: zelda_adventure_filesystem.ResourceFileSystemFolder):
        colorStream = StructStream(subFile.getRecord(0, kind="data"), endianPrefix=">")
        self.backgroundInitialColors = [colorStream.take("3B") for _ in range(240)]
        self.background = dyuvToRGB(subFile.getRecord(0, kind="video"), 384, 240, self.backgroundInitialColors)
    
    def _parseCollisionData(self, subFile: zelda_adventure_filesystem.ResourceFileSystemFolder):
        collisionMap = subFile.getRecord(1, kind="video")
        # The 4 index comes from code.
        self.collisionImage = rl7ToRGB(collisionMap, self.palette, emptySpaceColorIndex=4)

    def showAll(self):
        self.showInfo()
        self.showActors()
        self.showSprites()
        self.showScripts()

    def showInfo(self):
        display(self.info.__dict__)
        for i, cycler in enumerate(self.info.cyclers):
            print("Cycler", i)
            display(cycler.__dict__)

    def showActors(self):
        for i, actor in enumerate(self.actors):
            print("Actor", i)
            display(actor.__dict__)
        for i, desc in enumerate(self.descriptions):
            print("Description", i)
            display(desc.__dict__)
    
    def showSprites(self):
        display(self.background)
        display(self.collisionImage)
        for i, description in enumerate(self.descriptions):
            for j, group in enumerate(description.groups):
                for k, sprite in enumerate(group.sprites):
                    print("ActorDescription", i, "Group", j, "Sprite", k)
                    display(sprite)
            if description.extraUnusedGroups != None:
                for j, group in enumerate(description.extraUnusedGroups):
                    for k, sprite in enumerate(group):
                        print("ActorDescription", i, "Unused Group", j, "Sprite", k)
                        display(sprite)
        for i, group in enumerate(self.unusedSprites):
            for j, sprite in enumerate(group):
                print("Unused sprite group", i, "Sprite", j)
                display(sprite)

    def showScripts(self):
        for i, desc in enumerate(self.descriptions):
            print("Desc ID", i)
            display(desc.scripts)
        display(self.scripts)
        print("Cell vars:", self.vars)
        
        if len(self.extraScriptData) > 0:
            print("Unknown extra script data:")
            display(self.extraScriptData)
    
    def unusualDataFlags(self):
        ret = []
        if len(self.extraScriptData) > 0:
            ret.append("Extra script data")
        if len(self.unusedSprites) > 0:
            ret.append("Unused srpites")
        for i, desc in enumerate(self.descriptions):
            if desc.extraUnusedGroups:
                ret.append("Unused group for description index {}".format(i))
        if len(self.vars) > 1:
            ret.append("Has more than one script var")
        return ret

    def writeDataToDisk(self, root = ""):
        if len(root) > 0 and root[-1] != "/":
            root += "/"
        
        def easyMkdir(path):
            try:
                os.mkdir(path)
            except FileExistsError:
                pass
        easyMkdir(root + self.name)

        folder = root + self.name + "/"

        self.background.save(folder + "background.png", "png")
        self.collisionImage.save(folder + "metadataImage.png", "png")

        def defaultSerializer(o):
            try:
                return o.__dict__
            except:
                raise Exception("Unable to serialize object: " + repr(o))

        castJson = {
            "actors": [a.toPlainDict() for a in self.actors],
            "descriptions": [d.toPlainDict() for d in self.descriptions],
        }
        with open(folder + "cast.json", "w") as f:
            json.dump(castJson, f, default=defaultSerializer)
        
        convertedPalette = []
        for i in range(0, len(self.palette), 3):
            convertedPalette.append({
                "r": self.palette[i],
                "g": self.palette[i + 1],
                "b": self.palette[i + 2]
            })
        cellJson = {
            "palette": convertedPalette,
            "DYUVInitialValues": self.backgroundInitialColors,
            "scripts": self.scripts,
            "paletteCycles": None
        }
        if self.info.hasPaletteCycling:
            cellJson["paletteCycles"] = self.info.cyclers
        with open(folder + "cell.json", "w") as f:
            json.dump(cellJson, f, default=defaultSerializer)

        easyMkdir(folder + "sprites")
        for i, description in enumerate(self.descriptions):
            assert isinstance(description, ActorDescription), (i, description)
            easyMkdir(folder + "sprites/desc{}".format(i))
            for j, group in enumerate(description.groups):
                assert isinstance(group, SpriteGroup), (i, description, j, group)
                easyMkdir(folder + "sprites/desc{}/group{}".format(i, j))
                for k, sprite in enumerate(group.sprites):
                    sprite.save(folder + "sprites/desc{}/group{}/sprite{}.png".format(i, j, k), "png")
        
        

cellName = "s723"
cellData = underFiles.subFiles[cellName]
cell = Cell(cellData, cellName)
```

```python
def findUniqueActors(world: Dict[str, Cell], isOverworld: bool, knownActors: dict = {}) -> dict:
    cell: Cell
    desc: ActorDescription
    for cell in world.values():
        for i, desc in enumerate(cell.descriptions):
            spriteHash = desc.hashOfSpriteGroups()
            location = {
                "cell": cell.name,
                "isOverworld": isOverworld,
                "descIndex": i
            }
            if spriteHash not in knownActors:
                knownActors[spriteHash] = {
                    "hash": spriteHash,
                    "name": None,
                    "mainLocation": location,
                    "spriteGroups": desc.groups,
                    "locations": []
                }
            knownActors[spriteHash]["locations"].append(location)
    return knownActors

uniqueActors = findUniqueActors(overworldCells, True)
uniqueActors = findUniqueActors(underworldCells, False, uniqueActors)

len(uniqueActors)
```

```python
# Type ? to show all sprites for the actor.
# Type : followed by a number to skip to that actor.
# Type > followed by a string to only visit actors whose names contain that string.
# Type nothing to keep the original name, or if there was no previous name,
#     assign a new name like "unknownActor5".

skipTo = -1
filterStr = ""
for i, entry in enumerate(uniqueActors.values()):
    if i < skipTo:
        continue

    if len(filterStr) > 0 and (entry["name"] == None or filterStr not in entry["name"]):
        continue

    print("Actor", i, "/", len(uniqueActors))
    print("Hash:", entry["hash"])
    if entry["name"] != None:
        print("Name:", entry["name"])
    else:
        print("No name.")
    print("Cells:", [l["cell"] for l in entry["locations"]])
    display(entry["spriteGroups"][0].sprites[0])
    newName = input("New name: ").strip()

    if newName == "?":
        for group in entry["spriteGroups"]:
            for sprite in group.sprites:
                display(sprite)
        newName = input("New name: ").strip()
    
    if len(newName) > 0 and newName[0] == ">":
        filterStr = newName[1:].strip()
        continue

    if len(newName) > 0 and newName[0] == ":":
        try:
            skipTo = int(newName[1:].strip())
            continue
        except:
            skipTo = -1
            newName = input("Skip failed. New name: ").strip()

    if newName == "":
        if entry["name"] == None:
            entry["name"] = "unknownActor" + str(i)
            print("Assigned name:", entry["name"])
    else:
        entry["name"] = newName

```

# Scripting

## Opcodes

0. Nop
1. `save[index] = parameter`
2. `save[index] += parameter`
3. `save[index] -= parameter`
4. `player->health -= ONE_HEART`
5. Switch:
    1. If `index` is true (nonzero), then increase max health by 1 heart and fully heal.
    2. Else if `parameter` is zero, set an unused global value to max health.
    3. Else set an unused global value to `parameter` hearts.
6. `spawnAndAnimate_maybe(this)`
7. `sparkleAndDespawn_maybe(this)` Show sparkle effect on top / in front of zelda, maybe?
8. `teleportPlayer(cellName=(char*) &save[index])` Teleport zelda to a cell. `index` is usually a pointer to the cell's local variable section.
9. `player->position->x = parameter`
10. `player->position->y = parameter`
11. `scriptReturnValue = parameter`. Then if `index` is nonzero, play a sound. Not sure which sound, yet.
12. Switch:
    1. If `index` is true, unknown effect (function 0x2cd0).
    2. Else `spawnAndAnimate(cast[parameter])`
13. Switch:
    1. If `index` is 0, then `playSoundEffect(source=this, soundId=parameter)`.
    2. Else `playSoundEffect(source=cast[index - 10], soundId=parameter)` For non-Actor scripts, an actor index **must** be used. I have no idea why there's a -10 yet.
14. `actor->spriteGroup = parameter`. **Only allowed in actor scripts.**
15. Unknown effect. **Not allowed in actor scripts.**
16. `ignorePlayerInputs_maybe = parameter`
17. `registerAsTreasureIntercepter(this)`
18. Unknown effect. **Not allowed in actor scripts.**
19. Switch:
    1. If `index` is true, `rupees += parameter`
    2. Else `rupees -= parameter`
20. Dim the screen to really dark, maybe?
21. Unknown effect related to the Actor `parameter`'s sprite / animation.
22. Unknown effect.
23. `player->isVisible = parameter`
24. Unknown effect.
25. Update the cell's effective X and Y coords to `(index, parameter)`. This doesn't affect anything inside the cell, but changes how the load zones at the edges of the screen work.
26. Play unknown sound. **Only allowed in actor scripts.**
27. Unknown effect.

```python
def animateCycler(cycler: CycleDefinition, image: PIL.Image.Image, originalPalette: bytes, path: str):
    usedColors = [p for count, p in image.getcolors()]
    #print(usedColors)
    atLeastOneUsedColor = False
    for i in range(cycler.start, cycler.start + cycler.length):
        if i in usedColors:
            atLeastOneUsedColor = True
            break
    if not atLeastOneUsedColor:
        #print("Skipping cycler; no pixels use the colors")
        return
    
    subPalette = [b'\0\0\0\0'] * 256

    forwardPaletteSwap = list(range(256))
    backwardPaletteSwap = list(range(256))
    allColorsAreBlank = True
    includedColors = []
    for i in range(cycler.start, cycler.start + cycler.length):
        originalColor = originalPalette[i * 3:(i + 1) * 3]
        if originalColor != b'\0\0\0':
            allColorsAreBlank = False
        includedColors.append(originalColor)
        subPalette[i] = originalColor + b'\xFF'
        forwardPaletteSwap[i] = i + 1
        backwardPaletteSwap[i] = i - 1
    forwardPaletteSwap[cycler.start + cycler.length] = cycler.start
    backwardPaletteSwap[cycler.start] = cycler.start + cycler.length

    if allColorsAreBlank:
        #print("Skipping cycler; all colors are blank")
        return
    
    #display(includedColors)

    image.putpalette(b''.join(subPalette), "RGBA")

    frames: List[PIL.Image.Image] = []
    assert cycler.mode in ["IncreaseOnly", "Oscillate"], cycler.mode
    #assert cycler.cycleDirection == "Increasing"
    #assert cycler.staggerMode == "None"
    

    def stepAnimation(swapList):
        frames.append(image.copy())
        temp = copy.copy(subPalette)
        for i in range(len(subPalette)):
            subPalette[swapList[i]] = temp[i]
        image.putpalette(b''.join(subPalette), "RGBA")
    
    staggerAfterFrame = None
    if cycler.mode == "Oscillate":
        assert cycler.staggerMode == "None"
        assert cycler.hasLooped == False

        offset = cycler.currentOffset
        isIncreasing = cycler.cycleDirection == "Increasing"
        if not isIncreasing:
            assert cycler.cycleDirection == "Decreasing"

        while len(frames) < cycler.length * 2 - 1:
            if isIncreasing:
                stepAnimation(forwardPaletteSwap)
                offset += 1
                if offset == cycler.length:
                    isIncreasing = False
            else:
                stepAnimation(backwardPaletteSwap)
                offset -= 1
                if offset == 0:
                    isIncreasing = True

    else:
        if cycler.cycleDirection == "Increasing":
            swap = forwardPaletteSwap
        else:
            swap = backwardPaletteSwap
        for i in range(cycler.length):
            stepAnimation(swap)
            if cycler.staggerMode != "None" and i + cycler.currentOffset == cycler.length:
                staggerAfterFrame = i
    
    try:
        os.makedirs(path, exist_ok=True)
    except FileExistsError:
        pass

    for i, image in enumerate(frames):
        #display(image)
        image.save("{}/{}.png".format(path, i))

    frames[0].save("{}/animated.gif".format(path), save_all=True,
        append_images=frames[1:], duration=len(frames) * cycler.delay * 3, loop=0)
    
    with open("{}/metadata.json".format(path), "w") as f:
        summary = {
            "timeBetweenFrames_exceptFirstFrame": cycler.delay,
            "durationOfFirstFrame": abs(cycler.timer - cycler.delay),
            "putRandomDelayAfterFrame": None,
            "putExtraRandomDelayAfter2ndFrame": cycler.staggerMode != "None" and cycler.hasLooped
        }
        if staggerAfterFrame != None:
            summary["putRandomDelayAfterFrame"] = staggerAfterFrame
        json.dump(summary, f, indent=2)

def animateCyclersForCell(cell: Cell, path: str):
    for i, cycler in enumerate(cell.info.cyclers):
        if not cycler.enabled or cycler.length == 0:
            continue
        subPath = path + "/{}/{}".format(cell.name, i)
        animateCycler(cycler, cell.collisionImage.copy(), cell.palette, subPath)

#animateCyclersForCell(overworldCells["e20"], "export/cycles/overworld")
#underworldCells["s316"].showAll()
animateCyclersForCell(underworldCells["s716"], "export/cycles/underworld")

```

```python

def printOverlappingCycles(cells):
    for cell in cells.values():
        for i, cycler1 in enumerate(cell.info.cyclers):
            if cycler1.length == 0:
                continue
            colors1 = list(range(cycler1.start, cycler1.start + cycler1.length))
            for j, cycler2 in enumerate(cell.info.cyclers):
                if i >= j:
                    continue
                if cycler2.length == 0:
                    continue
                for color2 in range(cycler2.start, cycler2.start + cycler2.length):
                    if color2 in colors1:
                        print(cell.name + ",", i, "overlaps", j)
                        break
print("Overworld:")
printOverlappingCycles(overworldCells)
print("Underworld:")
printOverlappingCycles(underworldCells)
```
